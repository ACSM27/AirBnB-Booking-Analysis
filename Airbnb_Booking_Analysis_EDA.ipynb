### Data Cleaning

```python
# Converting the Data types of the columns
df['last review']=pd.to_datetime(df['last review'])
```

```python
# Checking if there are any missing values
df.isnull().sum()
```

```python
# Handling the Missing Values
df.fillna({'reviews per month': 0, 'last review': df['last review'].min()}, inplace=True)
```

```python
# Droping the values
df.dropna(subset=['NAME','host name'],inplace=True)
```

```python
# Remove dollar signs and convert to float
df['price'] = df['price'].replace('[\$,]', '',regex=True).astype(float)
df['service fee'] = df['service fee'].replace('[\$,]', '',regex=True).astype(float)
```

```python
# Dropping Duplicates
df.drop_duplicates(inplace=True)
```

```python
# Printing the information of all the data after removing missing and duplicate values
df.info()
```

### Visualizations


```python
# Distribution of Prices
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=30, kde=True)
plt.title('Distribution of Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

It shows a fairly even distribution of listing prices across different price ranges, indicating no particular concentration of listings in any specific price range. The KDE line helps visualize this even spread more clearly, confirming that the dataset contains listings with a wide variety of prices.

2. Room type Distribution

```python
# Room type distribution
sns.countplot(data=df, x='room type',color='blue',edgecolor='black')
plt.xticks(rotation=45)
plt.title('Room Type Distribution')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.show()
```

- It shows a clear distribution of the different room types available in the Airbnb dataset. The majority of listings are for â€˜Entire home/aptâ€™ and â€˜Private roomâ€™, with â€˜Shared roomâ€™ and â€˜Hotel roomâ€™ being much less common.

3. Neighbourhood Type Distribution

```python
# Distribution acrosss Neighbours
sns.countplot(data=df, x='neighbourhood group',color='blue',edgecolor='black')
plt.xticks(rotation=45)
plt.title('Neighbourhood Type Distribution')
plt.xlabel('Neighbourhood Type')
plt.ylabel('Count')
```

- It shows a clear distribution of the number of listings across different neighborhood groups. Manhattan and Brooklyn dominate the listings, suggesting they are prime locations for Airbnb. Queens, Bronx, and Staten Island have fewer listings, indicating less availability or popularity.

**4.  Price vs. Room Type**

Visualize the relationship between price and room type.

```python
# Price vs Room

sns.barplot(data=df,x='room type',y='price',hue='room type')
plt.xticks(rotation=45)
plt.title('Price vs. Room Type')
plt.xlabel("Room Type")
plt.ylabel("Price Type")
plt.show()
```

- It provides a detailed view of how prices vary across different room types in the Airbnb dataset. It shows that while â€˜Shared roomâ€™ tends to have lower prices, â€˜Private roomâ€™, â€˜Entire home/aptâ€™, and â€˜Hotel roomâ€™ have higher and more varied price ranges.

**5 . Reviews over time**

```python
# Reviews
reviews_over_time = df.groupby(df['last review'].dt.to_period('M')).size()
reviews_over_time.plot(kind='line',color='red')
plt.title('Number of Reviews Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.show()
```
